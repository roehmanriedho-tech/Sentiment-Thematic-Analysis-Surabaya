````python
import pandas as pd
import numpy as np
import warnings
from pathlib import Path
import json
from typing import Dict, List, Tuple, Optional
from dataclasses import dataclass, asdict
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

# ============================================================================
# CONFIGURATION AND CONSTANTS
# ============================================================================

@dataclass
class AnalysisConfig:
    """Configuration parameters for the analysis pipeline."""
    random_state: int = RANDOM_SEED
    cv_folds: int = 5
    n_permutations: int = 1000
    min_cohen_kappa: float = 0.67
    confidence_level: float = 0.95
    test_size: float = 0.2  # For initial exploration
    max_iter: int = 1000
    n_topics_range: Tuple[int, int] = (3, 10)

    def to_dict(self) -> Dict:
        """Convert config to dictionary for reporting."""
        return asdict(self)

config = AnalysisConfig()

# ============================================================================
# 1. DATA LOADING AND INITIAL EXPLORATION
# ============================================================================

class DataLoader:
    """Load and perform initial exploration of Instagram comments data."""

    def __init__(self, filepath: str):
        self.filepath = filepath
        self.df = None
        self.metadata = {}

    def load_data(self) -> pd.DataFrame:
        """Load CSV data with proper encoding handling."""
        print("=" * 80)
        print("LOADING DATA")
        print("=" * 80)

        try:
            # Try different encodings and delimiters
            potential_seps = [',', ';', '\t', '|']
            load_successful = False

            for encoding in ['utf-8', 'latin-1', 'cp1252']:
                for sep in potential_seps:
                    try:
                        self.df = pd.read_csv(
                            self.filepath,
                            encoding=encoding,
                            sep=sep
                        )
                        # Check if it looks like a valid dataframe (more than one column)
                        if len(self.df.columns) > 1:
                            print(f"‚úì Successfully loaded data with '{sep}' delimiter and {encoding} encoding")
                            load_successful = True
                            break # Exit sep loop
                        else:
                            # If only one column, it's likely the wrong delimiter
                            self.df = None # Reset df to try next combination
                    except (UnicodeDecodeError, pd.errors.ParserError):
                        continue
                if load_successful:
                    break # Exit encoding loop

            if self.df is None:
                raise ValueError("Could not load data with any combination of encoding and delimiter")

            # Store metadata
            self.metadata = {
                'total_comments': len(self.df),
                'columns': list(self.df.columns),
                'data_sources': self.df['Post Source'].unique().tolist(),
                'load_timestamp': datetime.now().isoformat()
            }

            print(f"\nDataset Overview:")
            print(f"- Total comments: {len(self.df)}")
            print(f"- Columns: {', '.join(self.df.columns)}")
            print(f"- Data sources: {', '.join(self.metadata['data_sources'])}")
            print(f"- Missing values: {self.df.isnull().sum().sum()}")

            # Display sample
            print(f"\nSample comments:")
            print(self.df[['Username', 'Comment', 'clean_text']].head(3).to_string())
            print()

            return self.df

        except Exception as e:
            print(f"‚úó Error loading data: {str(e)}")
            raise

    def validate_population_size(self, expected_n: int = 138) -> bool:
        """Validate that we have the expected population size."""
        actual_n = len(self.df)
        if actual_n != expected_n:
            print(f"‚ö† Warning: Expected N={expected_n}, but got N={actual_n}")
            return False
        print(f"‚úì Population size validated: N={actual_n}")
        return True

# ============================================================================
# 2. LANGUAGE DETECTION FOR CODE-MIXED CONTENT
# ============================================================================

class LanguageDetector:
    """
    Detect and handle code-mixed Indonesian-Javanese content.

    Note: This uses heuristic-based detection since standard language
    detection libraries (langdetect, langid) struggle with:
    - Code-mixing (Indonesian + Javanese)
    - Informal social media language
    - Regional dialects and slang
    """

    def __init__(self):
        # Javanese markers (common Javanese words)
        self.javanese_markers = [
            'ancen', 'mosok', 'wedi', 'wes', 'opo', 'koq', 'loh',
            'iku', 'iki', 'kuwi', 'kebacut', 'gak', 'ora', 'nek',
            'sing', 'karo', 'wingi', 'saiki', 'besok', 'esok'
        ]

        # Indonesian formal markers
        self.indonesian_markers = [
            'dengan', 'untuk', 'dari', 'yang', 'adalah', 'akan',
            'harus', 'dapat', 'oleh', 'pada', 'dalam', 'tentang'
        ]

    def detect_language_mix(self, text: str) -> Dict[str, any]:
        """
        Detect language composition in text.

        Returns:
            Dictionary with language features and classification
        """
        if pd.isna(text) or not text:
            return {
                'primary_language': 'unknown',
                'is_code_mixed': False,
                'javanese_score': 0.0,
                'indonesian_score': 0.0,
                'confidence': 0.0
            }

        text_lower = text.lower()
        words = text_lower.split()

        if not words:
            return {
                'primary_language': 'unknown',
                'is_code_mixed': False,
                'javanese_score': 0.0,
                'indonesian_score': 0.0,
                'confidence': 0.0
            }

        # Count markers
        javanese_count = sum(1 for word in words if word in self.javanese_markers)
        indonesian_count = sum(1 for word in words if word in self.indonesian_markers)

        # Calculate scores
        javanese_score = javanese_count / len(words)
        indonesian_score = indonesian_count / len(words)

        # Determine primary language and code-mixing
        if javanese_score > indonesian_score:
            primary = 'javanese'
            confidence = javanese_score
        elif indonesian_score > javanese_score:
            primary = 'indonesian'
            confidence = indonesian_score
        else:
            primary = 'mixed'
            confidence = (javanese_score + indonesian_score) / 2

        is_code_mixed = (javanese_score > 0 and indonesian_count > 0) or \
                       (javanese_score > 0.1 or indonesian_score > 0.1)

        return {
            'primary_language': primary,
            'is_code_mixed': is_code_mixed,
            'javanese_score': javanese_score,
            'indonesian_score': indonesian_score,
            'confidence': confidence
        }

    def analyze_dataset(self, df: pd.DataFrame, text_column: str = 'clean_text') -> pd.DataFrame:
        """Analyze language composition of entire dataset."""
        print("=" * 80)
        print("LANGUAGE DETECTION ANALYSIS")
        print("=" * 80)

        # Apply detection
        language_features = df[text_column].apply(self.detect_language_mix)

        # Create language feature columns
        for feature in ['primary_language', 'is_code_mixed', 'javanese_score',
                       'indonesian_score', 'confidence']:
            df[f'lang_{feature}'] = language_features.apply(lambda x: x[feature])

        # Report statistics
        print(f"\nLanguage Composition (N={len(df)}):")
        print(f"- Pure Indonesian: {(df['lang_primary_language'] == 'indonesian').sum()} "
              f"({(df['lang_primary_language'] == 'indonesian').sum()/len(df)*100:.1f}%)")
        print(f"- Pure Javanese: {(df['lang_primary_language'] == 'javanese').sum()} "
              f"({(df['lang_primary_language'] == 'javanese').sum()/len(df)*100:.1f}%)")
        print(f"- Code-mixed: {df['lang_is_code_mixed'].sum()} "
              f"({df['lang_is_code_mixed'].sum()/len(df)*100:.1f}%)")
        print(f"- Average Javanese score: {df['lang_javanese_score'].mean():.3f}")
        print(f"- Average Indonesian score: {df['lang_indonesian_score'].mean():.3f}")

        print(f"\n‚ö† LIMITATION: Heuristic-based detection may underestimate code-mixing")
        print(f"   due to informal language and regional variations.")
        print()

        return df

# ============================================================================
# 3. MANUAL ANNOTATION INTERFACE AND INTER-ANNOTATOR RELIABILITY
# ============================================================================

class AnnotationManager:
    """
    Manage manual annotations and calculate inter-annotator reliability.

    This implements a rigorous annotation framework with:
    - Multiple annotators
    - Cohen's Œ∫ for pairwise agreement
    - Krippendorff's Œ± for multiple annotators
    - Disagreement tracking and resolution
    """

    def __init__(self, df: pd.DataFrame):
        self.df = df
        self.annotations = {}
        self.agreement_metrics = {}

    def create_annotation_template(self, output_file: str = 'annotation_template.csv'):
        """Create template for manual annotation."""
        print("=" * 80)
        print("CREATING ANNOTATION TEMPLATE")
        print("=" * 80)

        template_df = self.df[['Username', 'Comment', 'clean_text']].copy()
        template_df['annotator_1_sentiment'] = ''
        template_df['annotator_1_sarcasm'] = ''
        template_df['annotator_1_theme'] = ''
        template_df['annotator_2_sentiment'] = ''
        template_df['annotator_2_sarcasm'] = ''
        template_df['annotator_2_theme'] = ''
        template_df['annotator_3_sentiment'] = ''
        template_df['annotator_3_sarcasm'] = ''
        template_df['annotator_3_theme'] = ''
        template_df['consensus_sentiment'] = ''
        template_df['consensus_sarcasm'] = ''
        template_df['consensus_theme'] = ''
        template_df['notes'] = ''

        template_df.to_csv(output_file, index=False)
        print(f"‚úì Annotation template created: {output_file}")
        print(f"  N = {len(template_df)} comments")
        print(f"\nAnnotation Guidelines:")
        print(f"  Sentiment: negative / neutral / positive")
        print(f"  Sarcasm: yes / no / uncertain")
        print(f"  Theme: [open coding, emerge from data]")
        print(f"\n‚ö† Ensure at least 3 independent annotators for robust reliability metrics")
        print()

        return output_file

    def load_annotations(self, annotation_file: str):
        """Load completed annotations with robust delimiter detection."""
        print(f"Attempting to load annotations from {annotation_file}...")

        # Try common delimiters with utf-8 encoding
        potential_seps = [',', ';', '\t', '|'] # Comma is default, try others
        for sep in potential_seps:
            for encoding in ['utf-8', 'latin-1']:
                try:
                    temp_df = pd.read_csv(annotation_file, sep=sep, encoding=encoding)
                    # Basic check: if only one column is read, and it's not explicitly delimited, it's likely wrong
                    # Our template has many columns.
                    if len(temp_df.columns) > 5: # Expecting more than 5 columns for proper annotation file
                        self.annotations = temp_df
                        print(f"‚úì Successfully loaded annotations from {annotation_file} with '{sep}' delimiter and '{encoding}' encoding.")
                        return self.annotations
                    else:
                        # print(f"  Attempt with separator '{sep}' and encoding '{encoding}' resulted in too few columns ({len(temp_df.columns)}). Trying other options.")
                        continue # Continue to next separator/encoding
                except Exception as e:
                    # print(f"  Attempt with separator '{sep}' and encoding '{encoding}' failed: {e}")
                    pass # Continue to next separator/encoding

        # If loop finishes without returning, raise an error
        self.annotations = {} # Ensure it's empty dict if all attempts fail
        raise ValueError(f"Could not load data from {annotation_file} with common delimiters (comma, semicolon, tab, pipe) and encodings (utf-8, latin-1). Please check the file format and ensure it's correctly delimited.")


    def calculate_cohens_kappa(self,
                               annotator1_col: str,
                               annotator2_col: str,
                               label: str = '') -> Dict:
        """
        Calculate Cohen's Œ∫ for pairwise inter-rater agreement.

        Target: Œ∫ ‚â• 0.67 (acceptable reliability)
        Interpretation:
        - Œ∫ < 0.40: Poor agreement
        - 0.40 ‚â§ Œ∫ < 0.60: Moderate agreement
        - 0.60 ‚â§ Œ∫ < 0.80: Substantial agreement
        - Œ∫ ‚â• 0.80: Almost perfect agreement
        """
        from sklearn.metrics import cohen_kappa_score

        # Remove missing values
        valid_mask = (self.annotations[annotator1_col].notna() &
                     self.annotations[annotator2_col].notna())

        if valid_mask.sum() == 0:
            return {
                'kappa': np.nan,
                'n': 0,
                'interpretation': 'No valid pairs',
                'meets_threshold': False
            }

        ann1 = self.annotations.loc[valid_mask, annotator1_col]
        ann2 = self.annotations.loc[valid_mask, annotator2_col]

        kappa = cohen_kappa_score(ann1, ann2)

        # Interpret
        if kappa < 0.40:
            interpretation = 'Poor agreement'
        elif kappa < 0.60:
            interpretation = 'Moderate agreement'
        elif kappa < 0.80:
            interpretation = 'Substantial agreement'
        else:
            interpretation = 'Almost perfect agreement'

        meets_threshold = kappa >= config.min_cohen_kappa

        result = {
            'kappa': kappa,
            'n': valid_mask.sum(),
            'interpretation': interpretation,
            'meets_threshold': meets_threshold,
            'label': label
        }

        return result

    def calculate_krippendorffs_alpha(self,
                                     annotator_cols: List[str],
                                     label: str = '') -> Dict:
        """
        Calculate Krippendorff's Œ± for multiple annotators.

        More robust than Cohen's Œ∫ for:
        - Multiple annotators (>2)
        - Missing data
        - Different sample sizes per annotator

        Interpretation similar to Cohen's Œ∫
        """
        try:
            import krippendorff
        except ImportError:
            print("‚ö† Installing krippendorff package...")
            import subprocess
            subprocess.check_call(['pip', 'install', 'krippendorff', '--quiet'])
            import krippendorff

        # Prepare reliability data matrix
        # Shape: (n_annotators, n_items)
        reliability_data = []

        for col in annotator_cols:
            # Convert to numeric codes
            values = self.annotations[col].values
            unique_values = [v for v in pd.unique(values) if pd.notna(v)]
            value_to_code = {v: i for i, v in enumerate(unique_values)}

            coded_values = []
            for v in values:
                if pd.isna(v):
                    coded_values.append(np.nan)
                else:
                    coded_values.append(value_to_code[v])

            reliability_data.append(coded_values)

        reliability_data = np.array(reliability_data)

        # Calculate Œ±
        alpha = krippendorff.alpha(reliability_data, level_of_measurement='nominal')

        # Interpret
        if alpha < 0.40:
            interpretation = 'Poor agreement'
        elif alpha < 0.60:
            interpretation = 'Moderate agreement'
        elif alpha < 0.80:
            interpretation = 'Substantial agreement'
        else:
            interpretation = 'Almost perfect agreement'

        meets_threshold = alpha >= config.min_cohen_kappa

        result = {
            'alpha': alpha,
            'n_annotators': len(annotator_cols),
            'n_items': reliability_data.shape[1],
            'interpretation': interpretation,
            'meets_threshold': meets_threshold,
            'label': label
        }

        return result

    def analyze_disagreements(self,
                             annotator_cols: List[str],
                             consensus_col: str,
                             text_col: str = 'clean_text') -> pd.DataFrame:
        """
        Identify and report disagreement cases.

        Returns:
            DataFrame with disagreement cases and their characteristics
        """
        # Find cases where annotators disagree
        disagreements = []

        for idx, row in self.annotations.iterrows():
            annotations = [row[col] for col in annotator_cols if pd.notna(row[col])]

            if len(annotations) < 2:
                continue

            # Check if all agree
            if len(set(annotations)) > 1:
                disagreements.append({
                    'index': idx,
                    'text': row[text_col],
                    'annotations': annotations,
                    'consensus': row[consensus_col] if pd.notna(row[consensus_col]) else 'unresolved',
                    'n_unique_labels': len(set(annotations))
                })

        disagreement_df = pd.DataFrame(disagreements)

        print(f"\nDisagreement Analysis:")
        print(f"- Total cases: {len(self.annotations)}")
        print(f"- Disagreement cases: {len(disagreements)} ({len(disagreements)/len(self.annotations)*100:.1f}%)")

        if len(disagreements) > 0:
            print(f"- Cases with 2 unique labels: {(disagreement_df['n_unique_labels'] == 2).sum()}")
            print(f"- Cases with 3+ unique labels: {(disagreement_df['n_unique_labels'] >= 3).sum()}")
            print(f"- Resolved: {(disagreement_df['consensus'] != 'unresolved').sum()}")
            print(f"- Unresolved: {(disagreement_df['consensus'] == 'unresolved').sum()}")

        return disagreement_df

    def calculate_full_reliability(self) -> Dict:
        """
        Calculate complete inter-annotator reliability metrics.

        Returns comprehensive reliability report.
        """
        print("=" * 80)
        print("INTER-ANNOTATOR RELIABILITY ANALYSIS")
        print("=" * 80)

        results = {
            'sentiment': {},
            'sarcasm': {},
            'metadata': {
                'n_total': len(self.annotations),
                'timestamp': datetime.now().isoformat(),
                'config': config.to_dict()
            }
        }

        # Sentiment reliability
        print("\n### SENTIMENT ANNOTATIONS ###")
        sentiment_annotators = [col for col in self.annotations.columns
                               if 'annotator' in col and 'sentiment' in col]

        if len(sentiment_annotators) >= 2:
            # Pairwise Cohen's Œ∫
            print("\nPairwise Cohen's Œ∫:")
            kappas = []
            for i in range(len(sentiment_annotators)):
                for j in range(i+1, len(sentiment_annotators)):
                    kappa_result = self.calculate_cohens_kappa(
                        sentiment_annotators[i],
                        sentiment_annotators[j],
                        label=f"{sentiment_annotators[i]} vs {sentiment_annotators[j]}"
                    )
                    kappas.append(kappa_result)
                    print(f"  {sentiment_annotators[i]} vs {sentiment_annotators[j]}: "
                          f"Œ∫ = {kappa_result['kappa']:.3f} ({kappa_result['interpretation']}) "
                          f"{'‚úì' if kappa_result['meets_threshold'] else '‚úó'}")

            results['sentiment']['pairwise_kappa'] = kappas

            # Krippendorff's Œ±
            if len(sentiment_annotators) >= 3:
                print("\nKrippendorff's Œ± (all annotators):")
                alpha_result = self.calculate_krippendorffs_alpha(
                    sentiment_annotators,
                    label='sentiment_all'
                )
                results['sentiment']['krippendorffs_alpha'] = alpha_result
                print(f"  Œ± = {alpha_result['alpha']:.3f} ({alpha_result['interpretation']}) "
                      f"{'‚úì' if alpha_result['meets_threshold'] else '‚úó'}")

            # Disagreement analysis
            disagreements = self.analyze_disagreements(
                sentiment_annotators,
                'consensus_sentiment'
            )
            results['sentiment']['disagreements'] = disagreements.to_dict('records')

        # Sarcasm reliability
        print("\n\n### SARCARSM ANNOTATIONS ###")
        sarcasm_annotators = [col for col in self.annotations.columns
                             if 'annotator' in col and 'sarcasm' in col]

        if len(sarcasm_annotators) >= 2:
            # Pairwise Cohen's Œ∫
            print("\nPairwise Cohen's Œ∫:")
            kappas = []
            for i in range(len(sarcasm_annotators)):
                for j in range(i+1, len(sarcasm_annotators)):
                    kappa_result = self.calculate_cohens_kappa(
                        sarcasm_annotators[i],
                        sarcasm_annotators[j],
                        label=f"{sarcasm_annotators[i]} vs {sarcasm_annotators[j]}"
                    )
                    kappas.append(kappa_result)
                    print(f"  {sarcasm_annotators[i]} vs {sarcasm_annotators[j]}: "
                          f"Œ∫ = {kappa_result['kappa']:.3f} ({kappa_result['interpretation']}) "
                          f"{'‚úì' if kappa_result['meets_threshold'] else '‚úó'}")

            results['sarcasm']['pairwise_kappa'] = kappas

            # Krippendorff's Œ±
            if len(sarcasm_annotators) >= 3:
                print("\nKrippendorff's Œ± (all annotators):")
                alpha_result = self.calculate_krippendorffs_alpha(
                    sarcasm_annotators,
                    label='sarcasm_all'
                )
                results['sarcasm']['krippendorffs_alpha'] = alpha_result
                print(f"  Œ± = {alpha_result['alpha']:.3f} ({alpha_result['interpretation']}) "
                      f"{'‚úì' if alpha_result['meets_threshold'] else '‚úó'}")

            # Disagreement analysis
            disagreements = self.analyze_disagreements(
                sarcasm_annotators,
                'consensus_sarcasm'
            )
            results['sarcasm']['disagreements'] = disagreements.to_dict('records')

        print("\n" + "=" * 80)
        print(f"‚ö† RELIABILITY THRESHOLD: Œ∫ ‚â• {config.min_cohen_kappa}")
        print(f"  This threshold represents 'substantial agreement' in the literature.")
        print(f"  Results below this threshold indicate need for:")
        print(f"  - Clarified coding guidelines")
        print(f"  - Additional annotator training")
        print(f"  - Discussion of ambiguous cases")
        print("=" * 80)
        print()

        self.agreement_metrics = results
        return results

# ============================================================================
# MAIN EXECUTION BLOCK
# ============================================================================

def main():
    """Main execution function for exploratory analysis."""

    print("\n")
    print("=" * 80)
    print("INSTAGRAM COMMENTS ANALYSIS - EXPLORATORY STUDY")
    print("Urban Transit Service Disruptions in Surabaya")
    print("=" * 80)
    print(f"Analysis Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"Random Seed: {RANDOM_SEED}")
    print(f"Configuration: {config.to_dict()}")
    print("=" * 80)
    print()

    # 1. Load Data
    loader = DataLoader('/content/sample_data/instagram_comments_analysis.csv') # Corrected file path
    df = loader.load_data()
    loader.validate_population_size(expected_n=138)

    # 2. Language Detection
    lang_detector = LanguageDetector()
    df = lang_detector.analyze_dataset(df)

    # 3. Create Annotation Template
    annotation_mgr = AnnotationManager(df)
    template_file = annotation_mgr.create_annotation_template()

    print("=" * 80)
    print("NEXT STEPS FOR COMPLETE ANALYSIS")
    print("=" * 80)
    print(f"\n1. MANUAL ANNOTATION (Required before proceeding):")
    print(f"   - File: {template_file}")
    print(f"   - Recruit 3 independent annotators")
    print(f"   - Provide clear coding guidelines")
    print(f"   - Annotate all {len(df)} comments")
    print(f"   - Focus on: sentiment, sarcasm, themes")
    print(f"\n2. INTER-ANNOTATOR RELIABILITY:")
    print(f"   - Load completed annotations")
    print(f"   - Calculate Cohen's Œ∫ and Krippendorff's Œ±")
    print(f"   - Resolve disagreements through discussion")
    print(f"   - Target: Œ∫ ‚â• {config.min_cohen_kappa}")
    print(f"\n3. MODEL DEVELOPMENT:")
    print(f"   - Use consensus labels as ground truth")
    print(f"   - Implement k-fold cross-validation")
    print(f"   - Calculate confidence intervals")
    print(f"   - Run permutation tests")
    print(f"\n4. METHODOLOGICAL CAVEATS:")
    print(f"   - Small sample size (N=138)")
    print(f"   - Code-mixed language challenges")
    print(f"   - Sarcasm detection in Indonesian context")
    print(f"   - Limited generalizability (single city, single event)")
    print("=" * 80)
    print()

    # Save initial analysis
    output_dir = Path('/content/analysis_outputs')
    output_dir.mkdir(exist_ok=True)

    # Save enriched dataset
    df.to_csv(output_dir / 'comments_with_language_features.csv', index=False)

    # Save metadata
    metadata = {
        'dataset': loader.metadata,
        'language_analysis': {
            'code_mixed_count': int(df['lang_is_code_mixed'].sum()),
            'code_mixed_percentage': float(df['lang_is_code_mixed'].sum() / len(df) * 100),
            'primary_languages': df['lang_primary_language'].value_counts().to_dict()
        },
        'configuration': config.to_dict(),
        'timestamp': datetime.now().isoformat()
    }

    with open(output_dir / 'analysis_metadata.json', 'w') as f:
        json.dump(metadata, f, indent=2)

    print(f"‚úì Analysis outputs saved to: {output_dir}")
    print()

    return df, annotation_mgr, metadata

if __name__ == "__main__":
    df, annotation_mgr, metadata = main()
    print("\n‚úì Phase 1 complete. Proceed with manual annotation before continuing.")
````

## Calculate Confidence Intervals

### Subtask:
Compute confidence intervals for the performance metrics obtained from the sentiment and sarcasm models to understand the variability and reliability of the model's performance.


**Reasoning**:
I will calculate the confidence intervals for the accuracy of both the sentiment and sarcasm models. Since the test set sizes are relatively small, I will use the Wilson score interval method for proportion confidence intervals, as it is more robust than the normal approximation for such cases. I will then print these intervals and provide a brief analysis.



````python
from statsmodels.stats.proportion import proportion_confint
import numpy as np

print("=" * 80)
print("CONFIDENCE INTERVALS FOR MODEL ACCURACY")
print("=" * 80)

# --- Sentiment Model Confidence Interval ---
# Calculate accuracy for sentiment model
accuracy_sentiment = accuracy_score(y_test_sentiment, y_pred_sentiment)
n_sentiment = len(y_test_sentiment)

# Calculate Wilson score interval for sentiment accuracy
# Using 95% confidence level (alpha=0.05)
conf_interval_sentiment = proportion_confint(
    count=round(accuracy_sentiment * n_sentiment),
    nobs=n_sentiment,
    alpha=1 - config.confidence_level,
    method='wilson'
)

print(f"\nSentiment Model Accuracy: {accuracy_sentiment:.4f}")
print(f"Sentiment Model 95% Confidence Interval (Wilson): ({conf_interval_sentiment[0]:.4f}, {conf_interval_sentiment[1]:.4f})")

# --- Sarcasm Model Confidence Interval ---
# Calculate accuracy for sarcasm model
accuracy_sarcasm = accuracy_score(y_test_sarcasm, y_pred_sarcasm)
n_sarcasm = len(y_test_sarcasm)

# Calculate Wilson score interval for sarcasm accuracy
# Using 95% confidence level (alpha=0.05)
conf_interval_sarcasm = proportion_confint(
    count=round(accuracy_sarcasm * n_sarcasm),
    nobs=n_sarcasm,
    alpha=1 - config.confidence_level,
    method='wilson'
)

print(f"\nSarcasm Model Accuracy: {accuracy_sarcasm:.4f}")
print(f"Sarcasm Model 95% Confidence Interval (Wilson): ({conf_interval_sarcasm[0]:.4f}, {conf_interval_sarcasm[1]:.4f})")

print("=" * 80)
````

**Sarcasm Model:**

Similarly, the sarcasm model's accuracy has a wide confidence interval. The class distribution for sarcasm was 'no' (99) and 'yes' (33), still imbalanced but less severely than sentiment. **Overall Reliability & Class Imbalance:**

Both models exhibit accuracies primarily driven by their ability to predict the majority class. The wide confidence intervals underscore that these point estimates of accuracy are not very precise, especially given the small test set. This lack of precision, combined with the poor performance on minority classes (as seen in the classification reports), suggests that while the overall accuracy might appear acceptable, the models are not reliable for predicting the less frequent categories (e.g., positive sentiment, sarcastic comments).

## Introduce Class Imbalance Strategies

### Subtask:
Explain the rationale behind addressing class imbalance and the specific strategies to be used (SMOTE for oversampling).

#### Rationale for Addressing Class Imbalance
Class imbalance occurs when the number of observations belonging to one class is significantly lower than those belonging to other classes. This is a common problem in many real-world datasets, including our sentiment and sarcasm detection tasks, where certain sentiments (e.g., 'positive' or 'neutral') or sarcasm labels ('yes') are far less frequent than others.

**Why is Class Imbalance a Problem?**
1.  **Biased Model Training:** Standard machine learning algorithms are often designed to maximize overall accuracy. When faced with imbalanced data, they tend to become biased towards the majority class, as correctly classifying the majority class most of the time leads to a high accuracy score, even if they completely ignore or misclassify the minority class.
2.  **Poor Minority Class Performance:** Consequently, models trained on imbalanced datasets often exhibit poor predictive performance (low precision, recall, and F1-score) on the minority class. This was clearly observed in our previous model evaluations, where the models failed to predict 'neutral' or 'positive' sentiment and 'yes' for sarcasm.
3.  **Misleading Evaluation Metrics:** Metrics like overall accuracy can be misleading in the presence of class imbalance. A model that predicts the majority class for all instances might achieve a high accuracy, but it would be useless for detecting the minority class, which is often the class of primary interest (e.g., detecting rare diseases, fraud, or, in our case, specific sentiments or sarcasm).

#### Oversampling as a Strategy
Oversampling is a technique used to address class imbalance by increasing the number of instances in the minority class. This helps to create a more balanced dataset, which can lead to better model performance, especially for the minority classes.

**Why Oversampling?**
*   It helps to provide the learning algorithm with more examples of the minority class, allowing it to learn the underlying patterns better.
*   Unlike undersampling (reducing the majority class), oversampling does not lead to a loss of potentially valuable information from the dataset.

#### Introducing SMOTE (Synthetic Minority Over-sampling Technique)
SMOTE is a popular and effective oversampling technique. Instead of simply duplicating existing minority class samples (which can lead to overfitting), SMOTE generates *synthetic* samples.

**How SMOTE Works:**
1.  For each sample in the minority class, SMOTE finds its k-nearest neighbors (samples that are closest in feature space).
2.  It then randomly selects one or more of these neighbors.
3.  It generates new synthetic samples along the line segments connecting the original sample to its randomly chosen neighbors. This involves taking the difference between the feature vectors of the original sample and its neighbor, multiplying it by a random number between 0 and 1, and adding it to the original sample's feature vector.

This approach creates new, distinct minority class samples that are similar to existing ones but not exact duplicates, thus expanding the decision boundary of the minority class without introducing exact copies.

#### Application Strategy
It is crucial to apply SMOTE **only to the training data**. Applying it to the entire dataset (including the test set) would lead to data leakage, where the model effectively 'sees' synthetic versions of the test data during training. This would result in overly optimistic and unrealistic performance evaluations.

#### Goal of Applying SMOTE
The primary goal of applying SMOTE is to improve the model's ability to detect and classify minority classes (e.g., 'neutral' and 'positive' sentiment, 'yes' sarcasm) accurately. By providing a more balanced training set, we aim to reduce the bias towards the majority class and achieve more reliable and generalizable models, without simply overfitting to the few existing minority samples. This should lead to better precision, recall, and F1-scores for the minority classes, while still maintaining reasonable overall performance.

### Penjelasan G*Power 3.1 dan Analisis Daya Post-hoc

G*Power 3.1 adalah program analisis daya statistik yang banyak digunakan dan gratis. Ini memungkinkan peneliti untuk menghitung ukuran sampel yang diperlukan (analisis *a priori*), atau kekuatan statistik dari pengujian yang telah dilakukan (analisis *post-hoc*).

**Analisis Daya Post-hoc** bertujuan untuk menentukan kekuatan pengujian yang telah dilakukan, *mengingat ukuran efek yang diamati, ukuran sampel, dan tingkat signifikansi alpha*. Jika hasil pengujian Anda tidak signifikan secara statistik (seperti yang ditunjukkan oleh nilai p-value yang tinggi dari tes permutasi), analisis daya *post-hoc* dapat membantu memahami apakah kurangnya signifikansi tersebut karena tidak ada efek yang sebenarnya, atau karena pengujian Anda memiliki daya yang terlalu rendah untuk mendeteksi efek yang ada (risiko kesalahan Tipe II yang tinggi).

Untuk regresi logistik, G*Power biasanya akan meminta parameter seperti:
*   Jenis Pengujian: Misalnya, $\chi^2$ - Goodness-of-fit, atau $\chi^2$ - Likelihood ratio test (untuk regresi logistik).
*   Ukuran Efek: Misalnya, rasio *odds* (odds ratio) atau perubahan $R^2$.
*   Ukuran Sampel: Jumlah total observasi.
*   Alpha ($\alpha$): Tingkat signifikansi (umumnya 0.05).
*   Derajat Kebebasan: Untuk model (misalnya, jumlah prediktor).

Karena G*Power tidak dapat dijalankan di sini, kita akan menggunakan pustaka Python `statsmodels` untuk melakukan perhitungan daya yang relevan dengan hasil klasifikasi Anda.

````python
from statsmodels.stats.power import GofChisquarePower
import numpy as np

print("=" * 80)
print("ANALISIS DAYA POST-HOC (TERHADAP AKURASI MODEL)")
print("=" * 80)

# --- Parameter untuk Analisis Daya ---
alpha = 0.05 # Tingkat signifikansi
n_test = len(y_test_sentiment) # Ukuran set pengujian (27 untuk kedua model)
# n_df = 1 # Derajat kebebasan untuk membandingkan proporsi (benar vs salah) - removed as it causes TypeError
baseline_proportion = 0.5 # Proporsi acak untuk tugas klasifikasi biner

# --- Analisis Daya Model Sentimen ---
observed_accuracy_sentiment = accuracy_score(y_test_sentiment, y_pred_sentiment_resampled)

# Hitung ukuran efek (Cohen's w) untuk perbandingan akurasi vs baseline
p_observed_sentiment = np.array([observed_accuracy_sentiment, 1 - observed_accuracy_sentiment])
p_expected_sentiment = np.array([baseline_proportion, 1 - baseline_proportion])
effect_size_sentiment = np.sqrt(np.sum((p_observed_sentiment - p_expected_sentiment)**2 / p_expected_sentiment))

power_analysis_sentiment = GofChisquarePower().solve_power(
    effect_size=effect_size_sentiment,
    nobs=n_test,
    alpha=alpha
)

print(f"\nModel Sentimen (Akurasi Teramati {observed_accuracy_sentiment:.4f}):")
print(f"  Ukuran Efek (Cohen's w): {effect_size_sentiment:.4f}")
print(f"  Daya (Power) Post-hoc: {power_analysis_sentiment:.4f}")

# --- Analisis Daya Model Sarkasme ---
observed_accuracy_sarcasm = accuracy_score(y_test_sarcasm, y_pred_sarcasm_resampled)

# Hitung ukuran efek (Cohen's w) untuk perbandingan akurasi vs baseline
p_observed_sarcasm = np.array([observed_accuracy_sarcasm, 1 - observed_accuracy_sarcasm])
p_expected_sarcasm = np.array([baseline_proportion, 1 - baseline_proportion])
effect_size_sarcasm = np.sqrt(np.sum((p_observed_sarcasm - p_expected_sarcasm)**2 / p_expected_sarcasm))

power_analysis_sarcasm = GofChisquarePower().solve_power(
    effect_size=effect_size_sarcasm,
    nobs=n_test,
    alpha=alpha
)

print(f"\nModel Sarkasme (Akurasi Teramati {observed_accuracy_sarcasm:.4f}):")
print(f"  Ukuran Efek (Cohen's w): {effect_size_sarcasm:.4f}")
print(f"  Daya (Power) Post-hoc: {power_analysis_sarcasm:.4f}")

print("=" * 80)
````

### Menghasilkan P-value untuk Berbagai Perbandingan (Chi-square Tests)

Untuk menerapkan koreksi `multipletests`, kita memerlukan serangkaian p-value dari berbagai uji hipotesis. Di sini, kita akan membuat p-value tersebut dengan melakukan uji `chi-square test of independence` untuk setiap kluster tematik, membandingkan distribusi sentimen dalam kluster tersebut dengan distribusi sentimen keseluruhan dalam dataset.

**Hipotesis yang diuji untuk setiap kluster tematik:**
*   **Hipotesis Nol (H0):** Distribusi sentimen dalam kluster tematik ini sama dengan distribusi sentimen keseluruhan.
*   **Hipotesis Alternatif (H1):** Distribusi sentimen dalam kluster tematik ini berbeda secara signifikan dari distribusi sentimen keseluruhan.

````python
from scipy.stats import chi2_contingency
from statsmodels.stats.multitest import multipletests
import pandas as pd
import numpy as np

print("=" * 80)
print("MENGHITUNG P-VALUE UNTUK BERBAGAI PERBANDINGAN SENTIMEN")
print("=" * 80)

# Pastikan plotting_df sudah bersih dari NaN di kolom yang relevan
# (Ini sudah dilakukan di cell ed949192, tetapi diulang untuk keamanan)
plotting_df_cleaned = annotations_df.dropna(subset=['annotator_1_theme', 'annotator_1_sentiment']).copy()
plotting_df_cleaned['annotator_1_theme'] = plotting_df_cleaned['annotator_1_theme'].replace('', np.nan)
plotting_df_cleaned.dropna(subset=['annotator_1_theme'], inplace=True)

# Dapatkan distribusi sentimen keseluruhan
overall_sentiment_counts = plotting_df_cleaned['annotator_1_sentiment'].value_counts()
overall_sentiment_total = overall_sentiment_counts.sum()

p_values_raw = []
themes = []

# Lakukan uji Chi-square untuk setiap kluster tematik
for theme in plotting_df_cleaned['annotator_1_theme'].unique():
    if pd.isna(theme): # Lewati jika tema adalah NaN
        continue

    # Distribusi sentimen untuk kluster tematik saat ini
    theme_sentiment_counts = plotting_df_cleaned[plotting_df_cleaned['annotator_1_theme'] == theme]['annotator_1_sentiment'].value_counts()

    # Buat tabel kontingensi
    # Kolom: Sentimen. Baris: Kluster vs Keseluruhan
    # Untuk melakukan uji chi2, kita perlu memastikan semua kategori sentimen ada di kedua distribusi
    combined_counts = pd.DataFrame({
        'Theme': theme_sentiment_counts,
        'Overall': overall_sentiment_counts
    }).fillna(0) # Isi NaN dengan 0 jika ada kategori sentimen yang tidak muncul di salah satu

    # Pastikan jumlah baris untuk 'Overall' disesuaikan agar proporsinya tetap benar terhadap total_overall - total_theme
    # Ini adalah pendekatan yang lebih tepat untuk menguji apakah distribusi tema berbeda dari 'sisanya'
    # Namun, untuk kesederhanaan dan mengikuti asumsi perbandingan 'theme vs overall_population', kita bisa menggunakan chi2_contingency
    # dengan tabel kontingensi langsung.

    # Pendekatan yang lebih sederhana: Uji apakah distribusi sentimen dalam tema tersebut berbeda dari distribusi keseluruhan.
    # Ini sering dilakukan dengan membandingkan counts, dan chi2_contingency akan menghitung expected counts.
    # Untuk ini, kita butuh tabel kontingensi 2xN (kluster vs sisa, dan sentimen).

    # Mari kita buat tabel kontingensi untuk kluster vs 'sisanya' (selain kluster ini)
    other_comments_df = plotting_df_cleaned[plotting_df_cleaned['annotator_1_theme'] != theme]
    other_sentiment_counts = other_comments_df['annotator_1_sentiment'].value_counts()

    # Gabungkan counts dari tema dan 'sisanya'
    contingency_table = pd.DataFrame({
        'Current_Theme': theme_sentiment_counts,
        'Other_Themes': other_sentiment_counts
    }).fillna(0).astype(int)

    # Lakukan uji chi-square hanya jika tabel tidak kosong dan memiliki variasi
    if not contingency_table.empty and (contingency_table.sum(axis=0) > 0).all() and (contingency_table.sum(axis=1) > 0).all():
        chi2, p_value, _, _ = chi2_contingency(contingency_table)
        p_values_raw.append(p_value)
        themes.append(theme)
    else:
        print(f"Skipping chi-square for theme '{theme}' due to insufficient data or lack of variation.")

print("\n--- P-value Mentah (Sebelum Koreksi) ---")
for i, p in enumerate(p_values_raw):
    print(f"Tema '{themes[i]}': p-value = {p:.4f}")

print("\n" + "=" * 80)
print("MENERAPKAN KOREKSI BENJAMINI-HOCHBERG FDR")
print("=" * 80)

# Aplikasikan koreksi Benjamini-Hochberg FDR
if p_values_raw:
    rejected, p_adjusted, _, _ = multipletests(p_values_raw,
                                               alpha=0.05,
                                               method='fdr_bh')

    print("\n--- P-value yang Disesuaikan (Benjamini-Hochberg FDR) ---")
    results = pd.DataFrame({
        'Theme': themes,
        'Original P-value': p_values_raw,
        'Adjusted P-value': p_adjusted,
        'Rejected_H0 (alpha=0.05)': rejected
    })
    results['Original P-value'] = results['Original P-value'].map('{:.4f}'.format)
    results['Adjusted P-value'] = results['Adjusted P-value'].map('{:.4f}'.format)
    display(results)

    print("\nInterpretation:")
    print("Kolom 'Rejected_H0 (alpha=0.05)' menunjukkan apakah hipotesis nol ditolak pada tingkat signifikansi alpha=0.05 setelah koreksi FDR.")
    print("Jika 'True', berarti distribusi sentimen dalam kluster tematik tersebut secara signifikan berbeda dari kluster lainnya.")
else:
    print("Tidak ada p-value yang dihasilkan untuk diterapkan koreksi.")

print("=" * 80)
````

### Menghitung Cram√©r's V (Ukuran Efek Chi-square)

Cram√©r's V adalah ukuran efek untuk uji chi-square. Ini menunjukkan kekuatan asosiasi antara dua variabel kategorikal, berkisar dari 0 (tidak ada asosiasi) hingga 1 (asosiasi sempurna). Semakin tinggi nilai V, semakin kuat hubungannya.

````python
from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd

print("=" * 80)
print("MENGHITUNG CRAM√âR'S V UNTUK SETIAP KUSTER TEMATIK")
print("=" * 80)

# Pastikan plotting_df_cleaned sudah didefinisikan dari langkah sebelumnya
# Ini penting karena Cram√©r's V menggunakan tabel kontingensi yang sama
plotting_df_cleaned = annotations_df.dropna(subset=['annotator_1_theme', 'annotator_1_sentiment']).copy()
plotting_df_cleaned['annotator_1_theme'] = plotting_df_cleaned['annotator_1_theme'].replace('', np.nan)
plotting_df_cleaned.dropna(subset=['annotator_1_theme'], inplace=True)

cramer_v_results = []

for theme in plotting_df_cleaned['annotator_1_theme'].unique():
    if pd.isna(theme): # Lewati jika tema adalah NaN
        continue

    theme_sentiment_counts = plotting_df_cleaned[plotting_df_cleaned['annotator_1_theme'] == theme]['annotator_1_sentiment'].value_counts()
    other_comments_df = plotting_df_cleaned[plotting_df_cleaned['annotator_1_theme'] != theme]
    other_sentiment_counts = other_comments_df['annotator_1_sentiment'].value_counts()

    contingency_table = pd.DataFrame({
        'Current_Theme': theme_sentiment_counts,
        'Other_Themes': other_sentiment_counts
    }).fillna(0).astype(int)

    if not contingency_table.empty and (contingency_table.sum(axis=0) > 0).all() and (contingency_table.sum(axis=1) > 0).all():
        chi2, p_value, dof, expected = chi2_contingency(contingency_table)

        n = contingency_table.sum().sum() # Total number of observations
        k = contingency_table.shape[0]     # Number of rows (sentiment categories)
        r = contingency_table.shape[1]     # Number of columns (Current_Theme, Other_Themes)

        # Cram√©r's V formula: sqrt(chi2 / (n * min(k-1, r-1)))
        # Note: Here k and r are used for rows and columns of the contingency table,
        # where rows are sentiment categories and columns are theme groups.
        # So min(k-1, r-1) becomes min(num_sentiment_categories-1, num_theme_groups-1)
        # num_sentiment_categories is 3 (negative, neutral, positive)
        # num_theme_groups is 2 (Current_Theme, Other_Themes)
        phi2 = chi2 / n
        cramer_v = np.sqrt(phi2 / min(k - 1, r - 1))

        cramer_v_results.append({
            'Theme': theme,
            'Chi-square': chi2,
            'P-value': p_value,
            'Cramer_V': cramer_v
        })
    else:
        print(f"Skipping Cram√©r's V for theme '{theme}' due to insufficient data or lack of variation.")

df_cramer_v = pd.DataFrame(cramer_v_results)

print("\nHasil Perhitungan Cram√©r's V:")
display(df_cramer_v.style.format({'Chi-square': '{:.2f}', 'P-value': '{:.4f}', 'Cramer_V': '{:.4f}'}))

print("\nInterpretation of Cram√©r's V:")
print("- 0.00-0.10: Negligible association")
print("- 0.10-0.20: Weak association")
print("- 0.20-0.40: Moderate association")
print("- 0.40-0.60: Relatively strong association")
print("- > 0.60: Strong association")
print("=" * 80)
````

## Perform Chi-square Test for Term-Sentiment Association

### Subtask:
Perform a chi-square test of independence for each term in the count matrix, comparing its frequency with the 'negative' and 'neutral' sentiment labels. This will identify terms whose frequencies are significantly associated with one of the sentiments.


**Reasoning**:
I will perform a chi-square test for each term to identify its statistical association with 'negative' versus 'neutral' sentiments. This involves iterating through each vectorized term, constructing a contingency table comparing term presence/absence with sentiment, and then calculating the chi-square statistic and p-value for each term.



````python
from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd

print("=" * 80)
print("CHI-SQUARE TEST FOR TERM-SENTIMENT ASSOCIATION")
print("=" * 80)

# 1. Get the feature names (terms) from the vectorizer
feature_names = vectorizer.get_feature_names_out()

# 2. Initialize empty lists
chi2_stats = []
p_values = []
terms = []

# Convert X_vectorized to a dense array for easier indexing
X_dense = X_vectorized.toarray()

# Get indices for negative and neutral sentiments
neg_indices = y_sentiment_filtered[y_sentiment_filtered == 'negative'].index
neu_indices = y_sentiment_filtered[y_sentiment_filtered == 'neutral'].index

# Get the original DataFrame indices corresponding to filtered sentiments
original_indices = X.index[mask] # Indices from the original X before filtering

# Map original indices to the filtered X_dense row indices
# This assumes X_filtered and y_sentiment_filtered maintain the order of X after masking
neg_rows = [i for i, idx in enumerate(original_indices) if y_sentiment_filtered.iloc[i] == 'negative']
neu_rows = [i for i, idx in enumerate(original_indices) if y_sentiment_filtered.iloc[i] == 'neutral']

# 3. Iterate through each feature (column) in X_vectorized
for i, feature in enumerate(feature_names):
    # a. Extract the counts for the current feature
    term_counts = X_dense[:, i]

    # b. Create a contingency table
    # Term present in negative sentiment comments
    term_in_neg = np.sum(term_counts[neg_rows] > 0)
    # Term present in neutral sentiment comments
    term_in_neu = np.sum(term_counts[neu_rows] > 0)

    # Term absent in negative sentiment comments
    term_not_in_neg = len(neg_rows) - term_in_neg
    # Term absent in neutral sentiment comments
    term_not_in_neu = len(neu_rows) - term_in_neu

    contingency_table = np.array([
        [term_in_neg, term_in_neu],
        [term_not_in_neg, term_not_in_neu]
    ])

    # c. Ensure the contingency table has valid dimensions and non-zero marginal sums
    if np.any(contingency_table.sum(axis=1) == 0) or np.any(contingency_table.sum(axis=0) == 0):
        chi2_stats.append(np.nan)
        p_values.append(np.nan)
        terms.append(feature)
    else:
        # d. Perform the chi-square test of independence
        chi2, p, dof, expected = chi2_contingency(contingency_table)

        # e. Append the calculated results
        chi2_stats.append(chi2)
        p_values.append(p)
        terms.append(feature)

# 4. Create a DataFrame named chi2_results_df
chi2_results_df = pd.DataFrame({
    'Term': terms,
    'Chi2_Statistic': chi2_stats,
    'P_value': p_values
})

chi2_results_df = chi2_results_df.sort_values(by='P_value', ascending=True).reset_index(drop=True)

print("\nTop 10 terms by Chi-square P-value (most significant associations):")
display(chi2_results_df.head(10))
print("\n" + "=" * 80)
````

## Identify and Analyze Top Associated Terms

### Subtask:
Filter terms based on significant p-values from the chi-square test. For these significant terms, calculate association scores to determine which sentiment is more dominantly related. Identify and display the top terms for each sentiment.


**Reasoning**:
To analyze the association of terms with sentiment, I will first filter the `chi2_results_df` for statistically significant terms. Then, for these terms, I will calculate the log-odds ratio to quantify their association with 'negative' versus 'neutral' sentiments. Finally, I will identify and display the top terms most strongly associated with each sentiment.



````python
import numpy as np
import pandas as pd

print("=" * 80)
print("IDENTIFYING TOP ASSOCIATED TERMS WITH SENTIMENTS")
print("=" * 80)

# 1. Set a significance level (alpha = 0.05)
alpha = 0.05
significant_terms_df = chi2_results_df[chi2_results_df['P_value'] < alpha].copy()

print(f"\nNumber of significant terms (p-value < {alpha}): {len(significant_terms_df)}")

if not significant_terms_df.empty:
    # Get the feature names (terms) from the vectorizer
    feature_names = vectorizer.get_feature_names_out()

    # Get indices for negative and neutral sentiments
    neg_indices_in_X_filtered = [i for i, label in enumerate(y_sentiment_filtered) if label == 'negative']
    neu_indices_in_X_filtered = [i for i, label in enumerate(y_sentiment_filtered) if label == 'neutral']

    # Calculate total word counts for each sentiment class
    total_words_neg = X_dense[neg_indices_in_X_filtered, :].sum() # Sum of all word counts in negative comments
    total_words_neu = X_dense[neu_indices_in_X_filtered, :].sum() # Sum of all word counts in neutral comments

    log_odds_ratios = []
    for idx, row in significant_terms_df.iterrows():
        term = row['Term']
        term_index = np.where(feature_names == term)[0][0]

        # Get term counts for the current significant term across all comments
        term_counts_in_docs = X_dense[:, term_index]

        # Count occurrences of the term in negative and neutral comments
        count_term_in_neg_docs = term_counts_in_docs[neg_indices_in_X_filtered].sum()
        count_term_in_neu_docs = term_counts_in_docs[neu_indices_in_X_filtered].sum()

        # Count occurrences of other words in negative and neutral comments (approximation)
        # Add a pseudo-count of 1 to all counts to avoid division by zero and log(0)
        pseudo_count = 1

        # Log-odds ratio calculation: log( (P(term|neg) / P(not term|neg)) / (P(term|neu) / P(not term|neu)) )
        # Approximated by: log( (count_term_in_neg / count_not_term_in_neg) / (count_term_in_neu / count_not_term_in_neu) )

        # Probability of term given negative sentiment
        p_term_given_neg = (count_term_in_neg_docs + pseudo_count) / (total_words_neg + X_dense.shape[1] * pseudo_count)
        # Probability of NOT term given negative sentiment
        p_not_term_given_neg = (total_words_neg - count_term_in_neg_docs + pseudo_count) / (total_words_neg + X_dense.shape[1] * pseudo_count)

        # Probability of term given neutral sentiment
        p_term_given_neu = (count_term_in_neu_docs + pseudo_count) / (total_words_neu + X_dense.shape[1] * pseudo_count)
        # Probability of NOT term given neutral sentiment
        p_not_term_given_neu = (total_words_neu - count_term_in_neu_docs + pseudo_count) / (total_words_neu + X_dense.shape[1] * pseudo_count)

        # Log-odds ratio
        if p_not_term_given_neg > 0 and p_not_term_given_neu > 0:
            log_odds = np.log(p_term_given_neg / p_not_term_given_neg) - np.log(p_term_given_neu / p_not_term_given_neu)
            log_odds_ratios.append(log_odds)
        else:
            log_odds_ratios.append(np.nan) # Should not happen with pseudo_count

    significant_terms_df['Log_Odds_Ratio'] = log_odds_ratios

    # Determine associated sentiment based on Log_Odds_Ratio
    # Positive log-odds: term is more associated with 'negative' sentiment
    # Negative log-odds: term is more associated with 'neutral' sentiment
    significant_terms_df['Associated_Sentiment'] = significant_terms_df['Log_Odds_Ratio'].apply(
        lambda x: 'negative' if x > 0 else ('neutral' if x < 0 else 'none')
    )

    significant_terms_df = significant_terms_df.dropna(subset=['Log_Odds_Ratio'])

    # 5. Identify and display the top 10 terms most strongly associated with 'negative' sentiment.
    top_negative_terms = significant_terms_df[significant_terms_df['Associated_Sentiment'] == 'negative']\
        .sort_values(by='Log_Odds_Ratio', ascending=False).head(10)

    print("\nTop Terms Associated with 'Negative' Sentiment (by Log-Odds Ratio):")
    display(top_negative_terms[['Term', 'Log_Odds_Ratio', 'P_value']])

    # 6. Identify and display the top 10 terms most strongly associated with 'neutral' sentiment.
    top_neutral_terms = significant_terms_df[significant_terms_df['Associated_Sentiment'] == 'neutral']\
        .sort_values(by='Log_Odds_Ratio', ascending=True).head(10) # Ascending for 'neutral' (more negative log-odds)

    print("\nTop Terms Associated with 'Neutral' Sentiment (by Log-Odds Ratio):")
    display(top_neutral_terms[['Term', 'Log_Odds_Ratio', 'P_value']])

else:
    print("No significant terms found at the specified alpha level.")

print("=" * 80)
````
